```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

df = pd.read_csv('data.csv')
afghanistan_data = df[df['Entity'] == 'Afghanistan']

data = afghanistan_data[['Year', 'Annual CO₂ emissions']].dropna()

X = data[['Year']] 
y = data['Annual CO₂ emissions']

model = DecisionTreeRegressor(random_state=42)
model.fit(X, y)

future_years = np.array([[2020], [2021], [2022], [2023], [2024], [2025]])

future_predictions = model.predict(future_years)

print("Predicted CO₂ emissions for the years 2020-2025:")
for year, prediction in zip(range(2020, 2026), future_predictions):
    print(f"Year {year}: {prediction:.2f} tons of CO₂")

plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, model.predict(X), color='red', label='Decision Tree Fit')

plt.scatter(future_years, future_predictions, color='green', label='Predictions (2020-2025)')

plt.xlabel('Year')
plt.ylabel('Annual CO₂ Emissions')
plt.legend()
plt.title('CO₂ Emissions Prediction using Decision Tree')
plt.show()
```
